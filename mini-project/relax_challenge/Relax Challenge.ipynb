Defining an  "adopted   user" as a user who has logged into the product on three separate days in at least one seven day period, identify which factors predict future user adoption  .

```python
# Count how many unique users in our engagment data of the 207,917 entries 
print(len(df1['user_id'].unique()))
```

```python
# Narrowing down users who logged in 3 times in total
users_three_logins = df1.groupby('user_id').filter(lambda X: len(X) >=3)
print(len(users_three_logins))
```

```python
print(len(users_three_logins['user_id'].unique()))
```

```python
users_three_logins.head()
```

```python
# Remove index
users_three_logins.reset_index(drop=True, inplace = True)

# Finding how many of the 2248 unique three-logins users are adopted users 
# who have logged in three separate days in at least one 7 day period 

adopted = 0
adopted_users = []
for i in range(len(users_three_logins)-2):
    user = users_three_logins['user_id'][i]
    if user != adopted and user == users_three_logins['user_id'][i+2]:
    # Get start date 
        sd = pd.Timestamp(users_three_logins['time_stamp'][i])
        # 7 days period end date
        ed = sd + pd.Timedelta('7D')
        if sd < pd.Timestamp(users_three_logins['time_stamp'][i+1]) < ed and sd < \
        pd.Timestamp(users_three_logins['time_stamp'][i+2]) < ed:
            adopted_users.append(user)
            adopted = user
```

```python
# Count how many adopted users there are 
len(adopted_users)
```

```python
# Add our new adopted_users feature to users dataframe: 1 for adopted and 0 for not-adopted

y = pd.Series(np.random.randn(len(df)))
n = 0
for i in range(len(df)):
    if df['object_id'][i] == adopted_users[n]:
        y[i] = 1
        n = n+1
        if n > len(adopted_users)-1:
            n = n -1
    else:
        y[i] = 0
        
users2 = pd.DataFrame(y, columns = ['adopted_users'])
df2 = pd.concat([df, users2], axis=1)
print((df2[df2['adopted_users'] == 1]).count())
```

```python
# there is also that some users were not invited by another user
# Construct another feature of invited: 1 for invited and 0 for wasn't invited 
inv = pd.Series(np.random.randn(len(df2)))
for i in range(len(df2)):
    if df['invited_by_user_id'][i] >=1:
        inv[i] = 1
    else:
        inv[i] = 0
df3 = pd.DataFrame(inv, columns = ['invited'])
df4 = pd.concat([df2,df3], axis = 1)
```
