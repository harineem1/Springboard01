```python
#Good transactions
df0 = df[df['Class'] == 0]
df0.shape
```

```python
#Fraud Transactions
df1 = df[df['Class'] == 1]
df1.shape
```

```python
#Plot the V1 vs V2 for good and fraud transactions
sns.regplot(x=df1["V1"], y=df1["V2"])
sns.regplot(x=df0["V1"], y=df0["V2"])
#Fraudulent transactions are concentrated at high values of v2 for a given v1
```

```python
#Plot the V1 vs V3 for good and fraud transactions
sns.regplot(x=df1["V1"], y=df1["V3"])
sns.regplot(x=df0["V1"], y=df0["V3"])
#Fraudulent transactions are concentrated at low values of v3 for a given v1
```

```python
#Plot the V1 vs V8 for good and fraud transactions
sns.regplot(x=df1["V1"], y=df1["V8"])
sns.regplot(x=df0["V1"], y=df0["V8"])
#Fraudulent transactions are concentrated at both high and low values of v8 for a given v1, but they seem to be concentrated in the mid values of v1
```

```python
sns.regplot(x=df1["V2"], y=df1["V3"])
sns.regplot(x=df0["V2"], y=df0["V3"])
there seems to be a linear relationship between v2 and v3 for fradulent transactions.
```

```python
sns.regplot(x=df1["V2"], y=df1["V4"])
sns.regplot(x=df0["V2"], y=df0["V4"])
#A combination of high V2 and high V4 seems to favor a fradulent transaction
```

```python
#Good transaction amounts against time. To see what amounts are normally good,
sns.regplot(x=df0["Time"], y=df0["Amount"])
```

```python
#Most fraud transactions seem to be of low value <1000
sns.regplot(x=df1["Time"], y=df1["Amount"], color = 'red')
```
