### How are these parameters distibuted?
Let us try to get the distribution of the parameters. They all seems to be normally distributed, with a skew towards the right.

```python
plt.figure(figsize=(12,8*4))
gs = gridspec.GridSpec(7, 4)
for i, cn in enumerate(df[df.iloc[:,1:29].columns]):
    ax = plt.subplot(gs[i])
    sns.kdeplot(df[cn].values, bw=2, label = "V1")
    ax.set_xlabel('')
    ax.set_title('Parameter: ' + str(cn))
plt.show()
```

Q3; How are these parameters correlated to each other? 
Created two dataframes one with good transactions and the other with fraudulent transactions.  Now let us see correlations and display the fraudulent ones differently.

```python
#Good transactions
df0 = df[df['Class'] == 0]
df0.shape
```

```python
#Fraud Transactions
df1 = df[df['Class'] == 1]
df1.shape
```

```python
plt.figure(figsize=(12,8*4))
gs = gridspec.GridSpec(7, 4)
for i, cn in enumerate(df[df.iloc[:,1:29].columns]):
    ax = plt.subplot(gs[i])
    sns.regplot(x=df1["V1"], y=df1[cn])
    sns.regplot(x=df0["V1"], y=df0[cn])
    ax.set_xlabel('')
    ax.set_title('Parameter: ' + str(cn))
plt.show()
```

Here is a sample of two other parameters correlation between them.

```python
sns.regplot(x=df1["V2"], y=df1["V3"])
sns.regplot(x=df0["V2"], y=df0["V3"])
there seems to be a linear relationship between v2 and v3 for fradulent transactions.
```

```python
sns.regplot(x=df1["V2"], y=df1["V4"])
sns.regplot(x=df0["V2"], y=df0["V4"])
#A combination of high V2 and high V4 seems to favor a fradulent transaction
```

```python
#Good transaction amounts against time. To see what amounts are normally good,
sns.regplot(x=df0["Time"], y=df0["Amount"])
```

```python
#Most fraud transactions seem to be of low value <1000
sns.regplot(x=df1["Time"], y=df1["Amount"], color = 'red')
```

What is the transaction amounts in different transaction classes like fradulent and non-fradulent

```python
sns.regplot(x=df["Class"], y=df["Amount"])
```

Looks like fradulent transactions are of lower denominations only.

Is there a specific time period when fraudulent transactions occur

```python

sns.lineplot(data=df0['Amount'], label = "Valid")
plot = sns.lineplot(data =  df1['Amount'], label = "Fraud")
plot.set(xlabel ='Transactions', ylabel ='Amount')
plt.show()
```

From the data available, it looks like the fraudulent transactions occur continuously, interspersed between the good transactions
